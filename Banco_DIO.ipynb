```python
menu = """
[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair
=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
    opcao = input(menu)

    if opcao == "d":
      saldo += float(input("Informe o valor do depósito: "))
      print("Depósito")

    elif opcao == "s":
      print("Saque")
      valorSaque = float(input("Informe o valor do saque: "))
      if saldo < valorSaque:
        print("Saldo insuficiente!")
      elif numero_saques >= LIMITE_SAQUES:
        print("Limite de saques excedido!")
      else:
        saldo -= valorSaque
        numero_saques+=1

    elif opcao == "e":
      extrato = saldo
      print(f"Extrato do dia: R$ {extrato:.2f}")

    elif opcao == "q":
        break

    else:
        print("Operação inválida! Por favor selecione novamente a opção desejada!")
```
